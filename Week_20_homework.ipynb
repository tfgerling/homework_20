## 3. Load the tokenized Paradise Lost from the Gutenberg Corpus in NLTK.* https://www.nltk.org/book/ch02.html . Stem or lemmatize the words and find counts. Select the top 20 words and create a histogram. Exclude stop words and make sure you are including words of all capitalizations in your count. If there are any meaningless “words” (“thus” and single letters, etc.) that are produced in your list or top words, alter your logic to exclude them. Specify why you chose stemming or lemmatization.

I chose stemming thinking that getting words to their base/root word would be most helpful when counting words.

```python
import nltk
from nltk.corpus import gutenberg
import re
from nltk.probability import FreqDist

gutenberg.fileids()
```

```python
paradise = gutenberg.words('milton-paradise.txt')
paradise
```

```python
paradise_sentences = gutenberg.sents('milton-paradise.txt')
paradise_sentences
```

```python
print(len(gutenberg.words('milton-paradise.txt')))
print(len(gutenberg.sents('milton-paradise.txt')))
```

```python
#stop words are common words that have low information value
sw = set(nltk.corpus.stopwords.words('english'))
```

```python
from nltk.stem.snowball import SnowballStemmer
stemmer = SnowballStemmer(language='english')
```

The additional words removed were thus, thi, thou, thee, us and may.

```python
fdist = FreqDist()

for sentence in paradise_sentences:
    filtered_list = [w for w in sentence if w.lower() not in sw]
    new_filtered_list = [w.lower() for w in filtered_list if w.isalnum()]
    new_word_list = [stemmer.stem(w) for w in new_filtered_list]
    for word in new_word_list:
        #dropping word if it's a digit instead
        if re.match(r'\d+',word):
            continue
        #dropping the words that I felt added no value
        elif word == 'thus' or word == 'thi' or word == 'thou' or word == 'thee' or word == 'us' or word == 'may':
            continue
        #making sure to lowercase the word before adding    
        elif re.match(r'\w+',word.lower()):
            fdist[word] += 1
    print(new_word_list)
```

```python
#taking note of the number of sentences in this book
len(paradise_sentences)
```

```python
for word in fdist.most_common(20):
    times = fdist[word]
    #print(" -- %s occurred %s times" % (word, times))
    print(word)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
## Saving the 20 most common words
top_20_fdist = fdist.most_common(20)

## Converting to series for easier plotting
top_20_fdist = pd.Series(dict(top_20_fdist))

## Setting figure, ax into variables
fig, ax = plt.subplots(figsize=(10,10))

# Thought the barplot would be an interesting chart to show
## Setting plot to horizontal for easy viewing + setting title + display  
bar_plot = sns.barplot(x=top_20_fdist.values, y=top_20_fdist.index, orient='h', ax=ax)
plt.xlabel('Frequency')
plt.title('Frequency Distribution')
plt.show();
```

```python
# Setting figure size
plt.subplots(figsize=(10,10))

# Setting up histogram to use top 20 group
plt.hist(top_20_fdist)

# Adding labels/title   
plt.xlabel('Frequency')
plt.ylabel('Word')
plt.title('Frequency Distribution')
plt.show()
```

## 4. Perform Vader Sentiment Analysis on the book. Find the 5 most negative, 5 most positive, and 5 most neutral sentences in Paradise Lost*. http://www.nltk.org/howto/sentiment.html . This may take a while to run, so you can always start with a small subset of the data (100 sentences) and then once your code works as expected, expand it to the whole book and let it run.



```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()
```

```python
# setting up sent_out as a global variable to be able to use it later
sent_out = []

def put_together(input_sent):
    ''' This function puts the sentences back together'''
    # the join command puts the strings together but it needs to be in a list so it was appended to the output
    for i, words in enumerate(input_sent):
        new_sent = ' '.join(words).replace('[ ','').replace(' ]','').replace(" ' ", "'").replace(' ,', ',').replace(' :',':').replace(' .','.').replace(' ;',';').replace(' ?','?').replace(' !','!').replace(' - ','-').replace('( ','(').replace(' )',')').replace("\'","'")
        sent_out.append([new_sent])
    return sent_out
```

```python
# calling the function to put the sentences back together
put_together(paradise_sentences)
```

```python
sentiments = []

# couldn't get analyzer to run on a list so converted each one to a string before using the analyzer
for x in range(0,len(sent_out)):
    string1=''
    for i in sent_out[x]:
        string1=string1+i
    #print(string1)
    sid = SentimentIntensityAnalyzer()
    sentiment = sid.polarity_scores(string1)
    #print(sentiment)
    compound = sentiment['compound']
    pos = sentiment['pos']
    neu = sentiment['neu']
    neg = sentiment['neg']
    sentiments.append({
        "text":string1,
        "compound":compound,
        "positive":pos,
        "negative":neg,
        "neutral":neu})
```

```python
# verifying the number of sentences still matches what the original had (at one point it didn't)
len(sent_out)
```

```python
# putting the sentiments data into a dataframe for easier manipulation

df=pd.DataFrame(sentiments)
```
